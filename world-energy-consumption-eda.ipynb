```python
energy_data = pd.read_csv('/kaggle/input/world-energy-consumption/World_Energy_By_Country_And_Region_1965_to_2023.csv')
energy_data.head()
```

```python
energy_data.info()
```

```python
# Converting energy consumption values to numeric and handling missing values
energy_data_cleaned = energy_data.copy()

# Dropping rows where 'Country' is NaN
energy_data_cleaned = energy_data_cleaned.dropna(subset=['Country'])

# Converting all year columns to numeric, replacing non-numeric values with NaN
energy_data_cleaned.iloc[:, 1:] = energy_data_cleaned.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

# Checking the cleaned data
energy_data_cleaned.info()
```

```python
energy_data_cleaned.head()
```

```python
import plotly.express as px

import plotly.graph_objects as go

# Summarizing global energy consumption over the years
global_energy_trend = energy_data_cleaned.iloc[:, 1:].sum()

# Global Energy Consumption Trend
global_energy_trend_fig = px.line(
    x=global_energy_trend.index,
    y=global_energy_trend.values,
    title='Global Energy Consumption Trend (1965-2023)',
    labels={'x': 'Year', 'y': 'Energy Consumption (in arbitrary units)'},
)
global_energy_trend_fig.update_layout(xaxis_title='Year', yaxis_title='Energy Consumption')
global_energy_trend_fig.show()
```

```python
# Extracting regions data
regions = energy_data_cleaned[energy_data_cleaned['Country'].str.contains('Total')]

# Regional Energy Consumption Trends
regional_energy_trend_fig = go.Figure()
for index, row in regions.iterrows():
    regional_energy_trend_fig.add_trace(
        go.Scatter(
            x=row.index[1:], y=row.values[1:], mode='lines+markers', name=row['Country']
        )
    )
regional_energy_trend_fig.update_layout(
    title='Regional Energy Consumption Trends (1965-2023)',
    xaxis_title='Year',
    yaxis_title='Energy Consumption',
    legend_title='Region'
)
regional_energy_trend_fig.show()
```

```python
# Extracting the top 10 energy-consuming countries for the latest year (2023)
top_countries_2023 = energy_data_cleaned[['Country', '2023']].dropna().sort_values(by='2023', ascending=False).head(10)

# Top 10 Energy-Consuming Countries in 2023
top_countries_2023_fig = px.bar(
    top_countries_2023,
    x='Country',
    y='2023',
    title='Top 10 Energy-Consuming Countries in 2023',
    labels={'2023': 'Energy Consumption'}
)
top_countries_2023_fig.update_layout(xaxis_title='Country', yaxis_title='Energy Consumption')
top_countries_2023_fig.show()
```

```python
# Extracting and reshaping the data for regions for comparative analysis
region_data_long = regions.melt(id_vars=['Country'], var_name='Year', value_name='Energy Consumption')

# Comparative Analysis of Energy Consumption Across Regions
comparative_analysis_fig = px.line(
    region_data_long,
    x='Year',
    y='Energy Consumption',
    color='Country',
    title='Comparative Analysis of Energy Consumption Across Regions (1965-2023)',
    labels={'Year': 'Year', 'Energy Consumption': 'Energy Consumption'}
)
comparative_analysis_fig.update_layout(legend_title='Region')
comparative_analysis_fig.show()
```

```python
# Selecting data for every decade starting from 1965
decade_years = ['1965', '1975', '1985', '1995', '2005', '2015', '2023']
decade_data = energy_data_cleaned[['Country'] + decade_years]

# Melting the dataframe to long format for easier plotting
decade_data_long = decade_data.melt(id_vars='Country', var_name='Year', value_name='Energy Consumption')

# Checking the first few rows of the long format dataframe
decade_data_long.head()
```

```python
# Plotly: Comparison of Energy Consumption Between Decades
comparison_fig = px.bar(
    decade_data_long,
    x='Country',
    y='Energy Consumption',
    color='Year',
    title='Comparison of Energy Consumption Between Decades',
    barmode='group',
    labels={'Energy Consumption': 'Energy Consumption'},
)
comparison_fig.update_layout(xaxis_title='Country', yaxis_title='Energy Consumption')
comparison_fig.show()
```

```python
# Plotly: Top Energy Consumers for Each Decade
top_countries_decades = []

for year in decade_years:
    top_countries = energy_data_cleaned[['Country', year]].dropna().sort_values(by=year, ascending=False).head(10)
    top_countries['Year'] = year
    top_countries.columns = ['Country', 'Energy Consumption', 'Year']
    top_countries_decades.append(top_countries)

top_countries_decades_df = pd.concat(top_countries_decades)

top_countries_fig = px.bar(
    top_countries_decades_df,
    x='Country',
    y='Energy Consumption',
    color='Year',
    title='Top Energy Consumers for Each Decade',
    barmode='group',
    labels={'Energy Consumption': 'Energy Consumption'}
)
top_countries_fig.update_layout(xaxis_title='Country', yaxis_title='Energy Consumption')
top_countries_fig.show()
```

```python
# Filter data for Asia Pacific
asia_pacific_data = energy_data_cleaned[energy_data_cleaned['Country'].str.contains('Asia Pacific', case=False, na=False)]

# Melting the dataframe to long format for easier plotting
asia_pacific_data_long = asia_pacific_data.melt(id_vars='Country', var_name='Year', value_name='Energy Consumption')

# Display the first few rows of the data for Asia Pacific
asia_pacific_data_long.head()
```

```python
# Plotting the energy consumption trend for Total Asia Pacific
total_asia_pacific = asia_pacific_data_long[asia_pacific_data_long['Country'] == 'Total Asia Pacific']

total_asia_pacific_fig = px.line(
    total_asia_pacific,
    x='Year',
    y='Energy Consumption',
    title='Energy Consumption Trend for Total Asia Pacific (1965-2023)',
    labels={'Energy Consumption': 'Energy Consumption'}
)
total_asia_pacific_fig.update_layout(xaxis_title='Year', yaxis_title='Energy Consumption')
total_asia_pacific_fig.show()
```

```python
# Yearly Breakdown for Total Asia Pacific
yearly_breakdown_fig = px.bar(
    total_asia_pacific,
    x='Year',
    y='Energy Consumption',
    title='Yearly Breakdown of Energy Consumption for Total Asia Pacific',
    labels={'Energy Consumption': 'Energy Consumption'}
)

yearly_breakdown_fig.update_layout(xaxis_title='Year', yaxis_title='Energy Consumption')

yearly_breakdown_fig.show()
```
